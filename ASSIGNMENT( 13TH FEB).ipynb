```python
# ques4
# The LookupError class in pyhton forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively
# KeyError -when key is not present in the dictionary
dict = {"name" : "anjali" , "stream" : "aiml" , "course" : "data science"}
try:
    print(dict["ph_number"])
except KeyError as e:
    print(e)
```

```python
# IndexError - when we try access an element from a list or tuple from an index that is not present in the list
li = [1,2,3,4,5]
try:
    print(li[10])
except IndexError as e:
    print(e)
```
